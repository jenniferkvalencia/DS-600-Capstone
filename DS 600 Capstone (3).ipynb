```python
#import libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
object_columns = phish.select_dtypes(include=['object'])
print(object_columns)
```

```python
#removing non-integer columns to calculate correlations
phish_int = phish.drop(columns = object_columns)
phish_int.head()
```

```python
plt.figure(figsize=(10, 7), dpi=200)
sns.heatmap(phish_int.corr(),
            cmap="coolwarm",
            linewidths=0.5, annot=True, fmt=".2f")
```

```python
#Too many variables for correlation map - need to separate top correlations
```

```python
top_corr = phish_int.corrwith(phish_int['label']).abs().sort_values(ascending=False)[:15].index.to_list()
```

```python
top_corr
```

```python
plt.figure(figsize=(10, 7), dpi=200)
sns.heatmap(phish_int[top_corr].corr(),
            cmap="coolwarm",
            linewidths=0.5, annot=True, fmt=".2f")
```

```python
#0 is phishing, 1 is legitimate
import matplotlib.pyplot as plt
ax = plt.subplots(1,1,figsize=(10,8))
sns.countplot(x='label',data=phish_int)
plt.title('Phishing vs. Legitimate URLs')
plt.show()
```

KMeans Clustering Model

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
```

```python
#Use top correlating variables, but drop target variable ('label')
X = phish_int[top_corr].drop(columns='label',axis=1)
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
X_train, X_test = train_test_split(X_scaled, test_size = 0.2, random_state = 42)
```

```python
#Using n_clusters =2 since this is a binary problem (phishing or legitimate)
kmeans = KMeans(n_clusters=2, init='k-means++', random_state = 42)
kmeans.fit(X_train)
```

```python
#Visualizing the clusters. Must use PCA since there are multiple features
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_train) 
labels=kmeans.labels_
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='coolwarm', s=50, alpha=0.5)
plt.title('Phishing Clusters')
plt.show()
```

```python
sil_score = silhouette_score(X_train, labels)
print(sil_score)
```
