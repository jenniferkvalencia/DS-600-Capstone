```python
#import libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
object_columns = phish.select_dtypes(include=['object'])
print(object_columns)
```

```python
#removing non-integer columns to calculate correlations
phish_int = phish.drop(columns = object_columns)
phish_int.head()
```

```python
corr_matrix = phish_int.corr()
threshold = 0.5
cols_to_keep = corr_matrix.columns[(corr_matrix.abs() >= threshold).any()]
filtered_corr_matrix = corr_matrix.loc[cols_to_keep,cols_to_keep]
```

```python
plt.figure(figsize=(10,7),dpi=200)
sns.heatmap(filtered_corr_matrix, cmap='coolwarm', linecolor = 'white', linewidths = 2, annot=True)
```
